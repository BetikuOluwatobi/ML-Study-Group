```python
train = train.drop(["id"], axis = 1) #Dropping the id from the train dataset as it will not help in any way in predicting the target
test_ids = test['id'] #Keeping the test ids in a separate file for final prediction
test = test.drop(["id"], axis = 1) #Again dropping the ids from the test dataset as they are not going to help in final prediction
labels = labels.drop(["id"], axis = 1) #Similarly dropping the ids from the train labels dataset
```

## Joining the train and test dataset so as to prevent doing the same operation on train and test dataset twice

```python
df = train.append(test, ignore_index = True)
```

```python
df.describe()
```

```python
#columns to drop
drop_cols = ['num_private', 'wpt_name', 'funder', 'public_meeting', 'region_code', 'district_code','extraction_type',
       'extraction_type_group', 'payment', 'quantity_group', 'source', 'waterpoint_type', 'scheme_name','subvillage','region', 'ward', 'lga',
            'recorded_by', 'basin']
```

```python
df = df.drop(drop_cols, axis = 1)#dropping columns that wont have any effect on the model
```

```python
df['year_recorded'] = pd.to_datetime(df['date_recorded']).dt.year#using only the year info

df['year_recorded'].fillna(df['year_recorded'].mean(), inplace = True)
df['year_recorded'].replace(0, df['year_recorded'].mean(), inplace = True)

```

```python
df['construction_year'].fillna(df['construction_year'].mean(), inplace = True)
df['construction_year'].replace(0, 2004, inplace = True)
df['operation_year'] = df['year_recorded'] - df['construction_year']
```

```python
ncols = ['year_recorded', 'date_recorded', 'construction_year']
df.drop(ncols, axis = 1, inplace = True)
```

```python
values = [0,1]
df['population'].replace(values, df['population'].mean(), inplace=True)
df['population'].fillna(df['population'].mean(), inplace=True)
```

```python
df['scheme_management'].value_counts()
```

```python
cols = ['installer','scheme_management', 'permit']
for col in cols:
    if df[col].value_counts().index[0]==0:
        df[col] = df[col].fillna(df[col].value_counts().index[1])
    else:
        df[col] = df[col].fillna(df[col].value_counts().index[0])
```

```python
df.info()#all missing values have been dealt with and columns that wont have any impact on the model removed
```

```python
train = df.iloc[:59400,:]
test = df.iloc[59400:,:]
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from catboost import Pool, CatBoostClassifier, cv
```

```python
X_train, X_val, y_train, y_val = train_test_split(train, labels, random_state=42)
```

```python
categorical_features_indices = np.where(train.dtypes != np.float)[0]
clf = CatBoostClassifier(random_seed=42, custom_metric='Accuracy')
clf.fit(X_train, y_train, cat_features=categorical_features_indices)
clf.score(X_val, y_val)
```

```python
test_predictions = clf.predict(X_val)
```

```python
submission = pd.DataFrame(test_predictions, columns=['status_group'])
submission.insert(0, 'id', test_ids)
submission.reset_index()
submission.dropna(how = 'any', inplace = True)
submission.to_csv('submission.csv', index = False)
```
