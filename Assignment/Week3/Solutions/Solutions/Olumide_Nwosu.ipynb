```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('raw_fighter_details.csv')
```

```python
df.head()
```

```python
df = df.fillna(axis=0, method='bfill')
df.columns = ['Name', 'Height', 'Weight', 'Reach', 'Stance', 'DOB']
df['Fighter ID'] = x_list = list(range(df.shape[0]))

df.set_index('Fighter ID')
```

```python
df['Weight'] = [int(x[:3]) if type(x) != int else x for x in df['Weight']]
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt_df = df[['Weight']]
plt_df.index = plt_df.index.map(int) 
plt_df.plot(figsize = (20, 15))
plt.title('Weight of all 3000+ fighters')
plt.ylabel('Weight in lbs')
plt.xlabel('Fighter ID')
plt.show()
```

```python
ax = plt_df.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title("Visualization of fighters' weights using area plot")
ax.set_ylabel('Weights in lbs')
ax.set_xlabel('Fighter id')
plt.show()
```

```python

df.plot(kind='scatter', x='Fighter ID', y='Weight', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```
