```python
#Import the the dataset to work  on 
import pandas as pd
import numpy as np
data_wk3s = pd.read_csv(r'D:\Python_only\Assignment\Week3\data.csv')
print(data_wk3s)
```

```python
#Now I want make R_fighter as my index
data_wk3s_new_index=data_wk3s.set_index('R_fighter')
print(data_wk3s_new_index.sample(20))
```

```python
#Let me start the data claeaning here by droping all the NaN
na_drop=data_wk3s_new_index.dropna(axis=0)
print(na_drop)
#I put head(50) because I want to work on 50 data as example 
#Meanwhile I just used that to show that I have the iddea of bar graph
```

```python
#Let me start grap plotiing here by referncing the columns to be used
import matplotlib as mpl
import matplotlib.pyplot as plt
#I will plot the bar graph of 
na_drop.head(50).plot(kind='bar', y='R_age', figsize=(20,10), color='purple')
plt.xlabel('Name')
plt.ylabel('B_age')
plt.title('Bar chart of any 50 R_fighters against their B_age')
plt.show
```

```python
#I am going to plot the pie graph of weight class of first 50 winners 
graph_data = na_drop.sort_values('R_wins', ascending=False).head(50)
#print(graph_data)
graph_data['weight_class'].value_counts().plot(kind='pie', figsize=(20,10))
plt.show()
```

```python
#For the histogram, I will show the R_age of the graph_data above, i.e of first 50  winners
graph_data['R_age'].value_counts().plot(kind='hist', color='green', figsize=(15,8))
plt.xlabel('R_age')
plt.show()
```

```python
#The same data will be used to plot line tyyoe of graph but this time, itis going to be on R_Stance column
graph_data['R_Stance'].value_counts(). plot(kind='line', color='violet', figsize=(12,5))
plt.xlabel('R_Stance')
plt.ylabel('Frequency')
plt.show()
```

```python
#Lastly for now, I want plot scatter graph
graph_data.plot(kind='scatter', x='R_age', y='R_wins', color='red', figsize=(15,5))
plt.show()
```

```python
#This first cell will deal with importation of the file alone
import numpy as np
import pandas as pd
week3 = pd.read_csv(r'D:\Python_only\Assignment\Week3\raw_fighter_dataset.csv')
print(week3)
```

```python
#Data cleaning start from here
#Let me first drop the NaN.I'm going to do this on the entire dataset
nan_drop=week3.dropna()
print(nan_drop)
```

```python
#Let me first change the index of my dataframe
#I want to use "fighter_name" as my index, thereby create a new dataframe
#to do the above, check whether the column you want to set as index is unique or not by typing next line below
#week3['fighter_name'].is_unique; This gives True. You can set it as index now!
my_useful = pd.DataFrame(nan_drop.set_index('fighter_name'))
print(my_useful)
#now I get a new dataframe called my_useful
```

```python
#Now let me begin to clean the data
#I'm going to clean column "Weight" and "Reach"
#Firstly, Weight
weight=my_useful['Weight'].str.extract(r'^(\d{3})', expand=True)
print(weight.sample(20))
```

```python
#secondly, Reach
reach = my_useful['Reach'].str.extract(r'^(\d{2})', expand=True)
print(reach.sample(20))
```

```python
#secondly, Reach
reach = my_useful['Reach'].str.extract(r'^(\d{2})', expand=True)
print(reach.sample(20))
```

```python
#Permit me to use this local and unprofessional codes to replace 'weight' and 'reach' in my_useful
#I want to form a new dataframe from which I'm going to plot gragp/chart required
my_useful['Reach'] = reach[0]
my_useful['Weight'] = weight[0]
for_plot = pd.DataFrame(my_useful)
print(for_plot)
```

```python
#now that the dataframe has been cleaned to this level, we can derive some values from it
#I start with bar chart. fighter_name against Reach
import matplotlib as mpl
import matplotlib.pyplot as plt
#I'm goint to use Stance data
for_plot['Stance'].value_counts().plot(kind='bar', x='Stance', color='violet')
plt.xlabel('Stance')
plt.title('Number of each stance in the dataset')
plt.show()
```

```python
#Histogram use the same data
for_plot['Stance'].value_counts().plot(kind='hist', x='Stance', color='blue')
plt.title('Number of each stance in the dataset')
print(for_plot['Stance'].value_counts())
plt.show()
```

```python
#represent the same thing with pie chart
for_plot['Stance'].value_counts().plot(kind='pie', figsize=(7,7))
plt.title('Number of each stance in the dataset')
print(for_plot['Stance'].value_counts())
plt.legend()
plt.show()
```

```python
#Same data for lineplot
for_plot['Stance'].value_counts().plot(kind='line', figsize=(7,7))
print(for_plot['Stance'].value_counts())
plt.legend()
plt.show()
```
